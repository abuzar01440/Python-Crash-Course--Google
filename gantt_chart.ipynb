```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

# Define tasks with (start_week, duration in weeks)
tasks = {
    "Data collection & preprocessing": (1, 2),
    "Data augmentation & baseline training": (3, 2),
    "Select best segmentation + SpineR + instance assignment": (5, 2),
    "RLMA + Feature extraction": (7, 2),
    "Classifier training & evaluation": (9, 1),
    "Fine-tune MedGemma + agents": (10, 2),
    "Integration & final reporting": (12, 1),
}

# Colors for tasks
colors = [
    "#4daf4a", "#377eb8", "#ff7f00",
    "#984ea3", "#e41a1c", "#a65628", "#f781bf"
]

fig, ax = plt.subplots(figsize=(10, 6))

# Draw tasks
for i, (task, (start, duration)) in enumerate(tasks.items()):
    ax.barh(i, duration, left=start, color=colors[i % len(colors)], edgecolor="black")
    ax.text(start + duration/2, i, task, va='center', ha='center', color="black", fontsize=9, fontweight="bold")


# Formatting
ax.set_yticks(range(len(tasks)))
ax.set_yticklabels([])
ax.set_xlabel("Weeks")
ax.set_xlim(0.5, 12.5)
ax.set_xticks(range(1, 13))
ax.invert_yaxis()
ax.set_title("3-Month Research Proposal Timeline (Gantt Chart)", fontsize=12, fontweight="bold")

plt.tight_layout()
plt.show()
```
